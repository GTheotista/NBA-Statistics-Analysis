In the dataset, it is observed that some players appear multiple times in a given month due to team changes. For example: Player: Precious Achiuwa.

Ideally, only one record should be retained per player per month, specifically the row where Tm == TOT, which aggregates their performance across all teams played during that period. To clean the data, duplicate player entries should be removed, keeping only the TOT row where available. This ensures accurate and non-redundant representation of player statistics.

**Data Cleaning: Multiple Records**

```python
transferred_df = pd.DataFrame()
    
for nba_data, month in zip([nba_nov, nba_dec, nba_jan], ['2023-11', '2023-12', '2024-01']):
    # Identify players who switched teams
    team_changes = nba_data.groupby(['Month', 'Player'])['Tm'].apply(list).reset_index()

    # Process players who have 'TOT' in their records
    players_with_tot = team_changes[team_changes['Tm'].apply(lambda x: 'TOT' in x)].copy()

    def extract_teams(team_list):
        teams = [team for team in team_list if team != 'TOT']
        if len(teams) == 1:
            return teams[0], teams[0]
        elif len(teams) > 1:
            return teams[0], teams[-1]
        return None, None

    players_with_tot[['Old Team', 'Current Team']] = players_with_tot['Tm'].apply(lambda x: pd.Series(extract_teams(x)))

    # Merge extracted teams back to main dataset
    nba_data = nba_data.merge(players_with_tot[['Month', 'Player', 'Old Team', 'Current Team']], on=['Month', 'Player'], how='left')

    # Keep only 'TOT' rows if a player has 'TOT' in their records
    nba_data = nba_data[~((nba_data['Player'].isin(players_with_tot['Player'])) & (nba_data['Tm'] != 'TOT'))]

    # Ensure Tm1 and Tm2 are correctly assigned only for 'TOT' rows
    nba_data['Old Team'].fillna(nba_data['Tm'], inplace=True)
    nba_data['Current Team'].fillna(nba_data['Tm'], inplace=True)
    
    transferred_df = pd.concat([transferred_df, players_with_tot[['Player', 'Month']]], ignore_index=True)

    # Save the modified dataframe back to its corresponding variable
    if month == '2023-11':
        nba_nov = nba_data
    elif month == '2023-12':
        nba_dec = nba_data
    elif month == '2024-01':
        nba_jan = nba_data
```

```python
# Concatenate the dataframes for November, December, and January
nba_data = pd.concat([nba_nov, nba_dec, nba_jan], ignore_index=True)

# Display the combined dataframe
nba_data
```

Specifically, focus on FG%, 3P%, 2P%, eFG%, and FT%, since these can be recalculated from other available data.

These formulas are based on standard definitions used in basketball analytics, particularly from sources like the NBA and Basketball Reference:

- Field Goal Percentage (FG%): This metric represents how efficiently a player makes field goals by dividing successful field goals by total attempts.

- Three-Point Percentage (3P%): Similar to FG%, but specifically for three-point shots.

- Two-Point Percentage (2P%): Measures efficiency for two-point shots.

- Effective Field Goal Percentage (eFG%): Adjusts for the fact that three-pointers are worth more than two-pointers.

- Free Throw Percentage (FT%): Measures accuracy from the free-throw line.

These formulas are widely used in basketball statistics and can be found in analytics references:

https://www.nba.com/stats/help/glossary

```python
# Fill missing shooting percentages using actual formulas
nba_data['FG%'] = nba_data['FG'] / nba_data['FGA']
nba_data['3P%'] = nba_data['3P'] / nba_data['3PA']
nba_data['2P%'] = nba_data['2P'] / nba_data['2PA']
nba_data['eFG%'] = (nba_data['FG'] + 0.5 * nba_data['3P']) / nba_data['FGA']
nba_data['FT%'] = nba_data['FT'] / nba_data['FTA']

# Handle cases where denominator is 0 to avoid NaN values
nba_data[['FG%', '3P%', '2P%', 'eFG%', 'FT%']] = nba_data[['FG%', '3P%', '2P%', 'eFG%', 'FT%']].fillna(0)
```

## Statistical Analysis

This following code is ANOVA test checks whether there are significant differences in average Total Points, Assists, and Rebounds across different months, with a p-value < 0.05 indicating a significant variation.

*Hypotesis Testing*
- Null Hypothesis (H0): There is no significant difference in the mean Total Points, Total Assists, or Total Rebounds across different months. This means team performance (in terms of these metrics) remains consistent over time.
- Alternative Hypothesis (H1): There is at least one significant difference in the mean Total Points, Total Assists, or Total Rebounds across months, indicating that team performance varies over time.

If the p-value < 0.05, we reject H0

```python
from scipy import stats

# ANOVA test for Total Points
anova_pts = stats.f_oneway(*[team_stats[team_stats['Month'] == month]['Total_PTS'] for month in team_stats['Month'].unique()])
print("ANOVA Test for Total Points across months:", anova_pts)
print("Reject H0" if anova_pts.pvalue < 0.05 else "Fail to reject H0")

# ANOVA test for Total Assists
anova_ast = stats.f_oneway(*[team_stats[team_stats['Month'] == month]['Total_AST'] for month in team_stats['Month'].unique()])
print("ANOVA Test for Total Assists across months:", anova_ast)
print("Reject H0" if anova_ast.pvalue < 0.05 else "Fail to reject H0")

# ANOVA test for Total Rebounds
anova_trb = stats.f_oneway(*[team_stats[team_stats['Month'] == month]['Total_TRB'] for month in team_stats['Month'].unique()])
print("ANOVA Test for Total Rebounds across months:", anova_trb)
print("Reject H0" if anova_trb.pvalue < 0.05 else "Fail to reject H0")
```

This means that there are significant differences in team performance across different months for Total Points, Total Assists, and Total Rebounds. In other words, team performance in these metrics is not consistent over time, and at least one month's average differs significantly from the others.

This linear regression analysis is to determine how well Total Assists (AST) and Total Rebounds (TRB) can predict Total Points (PTS) for teams.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Features (Total Assists and Total Rebounds) and Target (Total Points)
X = team_stats[['Total_AST', 'Total_TRB']]
y = team_stats['Total_PTS']

# Linear Regression Model
model = LinearRegression()
model.fit(X, y)

# R-squared (measuring how well the model explains the variability in Total Points)
r2 = model.score(X, y)

print(f"R-squared value: {r2:.2f}")
print("Reject H0" if r2 > 0.1 else "Fail to reject H0")  # Threshold of 0.1 for significance
```

The R-squared value (0.99) indicates that 99% of the variability in Total Points can be explained by Total Assists and Total Rebounds using this linear regression model.

Since the R-squared value is much higher than 0.1, we reject H₀, meaning that there is a significant relationship between Total Points and the predictor variables (Total Assists and Total Rebounds). In other words, these features are strong predictors of a team's Total Points.

This following code is about linear regression for each NBA team to analyze the trend of total points scored per month. The results are plotted to visualize whether a team's scoring is increasing, decreasing, or stable over time, with R² values indicating how well the trend explains the variation in total points.

```python
# Encode Month as numerical (1 = Month 1, 2 = Month 2, etc.)
team_stats['Month_Num'] = team_stats['Month'].astype('category').cat.codes + 1

# Loop through each team
teams = team_stats['Current Team'].unique()
plt.figure(figsize=(12, 7))

for team in teams:
    team_data = team_stats[team_stats['Current Team'] == team]
    
    # Features and target variable
    X = team_data[['Month_Num']].values  # Feature (month)
    y = team_data['Total_PTS'].values  # Target (Total Points)
    
    # Linear Regression
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    
    # Plot regression results
    plt.plot(team_data['Month_Num'], y, 'o-', label=f"{team} (R²={r2_score(y, y_pred):.2f})")

plt.xlabel("Month")
plt.ylabel("Total Points")
plt.title("Total Points Trend per Month (Linear Regression) for All Teams")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

This following code predicts the total points for the next month for the top 5 NBA teams based on their total points in January 2024. It applies linear regression to each team's historical monthly total points and forecasts their total points for the following month. The predictions are stored in a DataFrame and displayed.

```python
# Select the top 5 teams based on Total_PTS in January
top_5_jan = team_stats[team_stats['Month'] == '2024-01'].nlargest(5, 'Total_PTS')

# Create a DataFrame to store predictions
predictions = []

# Loop through each team in the top 5 from January
for team in top_5_jan['Current Team'].values:
    team_data = team_stats[team_stats['Current Team'] == team]
    
    # Features and target variable
    X = team_data[['Month_Num']].values  # Ensure it's in 2D array format
    y = team_data['Total_PTS'].values  # Target (Total Points)
    
    # Linear Regression
    model = LinearRegression()
    model.fit(X, y)
    
    # Predict total points for the next month
    next_month = np.array([[X.max() + 1]])  # Next month (ensure 2D format)
    predicted_pts = model.predict(next_month)[0]  # Prediction
    
    # Store prediction results
    predictions.append({'Current Team': team, 'Month_Num': next_month[0, 0], 'Predicted_Total_PTS': round(predicted_pts)})

# Convert predictions to a DataFrame
predictions_df = pd.DataFrame(predictions)

# Display predicted results
print(predictions_df)
```

Objective: Identify which players are improving or declining over time.
Method: Fit a linear regression model for each player’s Total PTS, AST, and TRB over months.

```python
print('Top 10 Player Trend Based Total Points')

player_trends = []

for player in nba_data['Player'].unique():
    player_data = nba_data[nba_data['Player'] == player]
    
    if len(player_data) >= 3:  # Ensure at least 3 months of data
        X = np.array(player_data['Month'].astype('category').cat.codes + 1).reshape(-1, 1)  
        y = player_data['Total_PTS'].values

        model = LinearRegression()
        model.fit(X, y)
        trend = model.coef_[0]  # Slope of regression line

        player_trends.append({'Player': player, 'Trend': trend})

# Convert to DataFrame
player_trends_df = pd.DataFrame(player_trends).sort_values(by='Trend', ascending=False)

# Top 10 improving players
print(player_trends_df.head(10))  
```
